```python
%matplotlib inline

import pandas as pd
from sklearn.cross_validation import train_test_split
import numpy as np
from sklearn.externals import joblib
import Recommenders
import Evaluation as Evaluation
```

```python
triplets_file='https://static.turi.com/datasets/millionsong/10000.txt'#Downloading Triplets file

songs_metadata_file='https://static.turi.com/datasets/millionsong/song_data.csv'#Downloading Metadata
                                                                                #file.
```

```python
song_df_1=pd.read_table(triplets_file,header=None)#reading triplets file in table.
```

```python
song_df_1.columns = ['user_id', 'song_id', 'listen_count']#Specifying column names.
print(song_df_1)
song_df_1.head()
```

```python
song_df_2 =  pd.read_csv(songs_metadata_file)# Reading songs_metadata file.
print(song_df_2)
```

```python
#Merging two files for input dataframe for the Recommendation.
song_df = pd.merge(song_df_1, song_df_2.drop_duplicates(['song_id']), on="song_id", how="left") 
print(song_df)
```

```python
len(song_df) 
```

```python
song_df=song_df.head(10000)#Data transformation-We are gonna select a subset of data that contains 10,000
print(song_df)
len(song_df) 
```

```python
#creating a new column Song by merging title and artist name.
song_df['song'] = song_df['title'].map(str) + " - " + song_df['artist_name']#map(str)-converts list items in 
                                                                   #string format.
print(song_df['song'])   
```

```python
song_grouped = song_df.groupby(['song']).agg({'listen_count': 'count'}).reset_index()#For grouping 
#song Column and aggregating data(Listen_count) as count,When we reset the index, the old index is added as a column,
#and a new sequential index is used
song_grouped
grouped_sum = song_grouped['listen_count'].sum()#computes sum of group values(Listen_count)
song_grouped['percentage']  = song_grouped['listen_count'].div(grouped_sum)*100#Percentage of listen_count
song_grouped.sort_values(['listen_count', 'song'], ascending = [0,1])
```

```python
users=song_df['user_id'].unique() #Unique users from the 10,000's dataset.
len(users) #365 unique users.
```

```python
songs=song_df['song'].unique()
len(songs) #5151 unique songs
```

```python
#Splitting of training and testing data by train_test_split.
train_data,test_data=train_test_split(song_df, test_size=0.20, random_state=0)
len(train_data)
len(test_data)
```

```python
print(train_data.head())
```

```python
#POPULARITY  BASED RECOMMENDATION(Not personalized).

#Creating an instanse of popularity based recommenders class(Acting as a black box) 
pm = Recommenders.popularity_recommender_py()
pm.create(train_data, 'user_id', 'song') #instance is applied on training data.
```

```python
#Doing predictions using popularity model
user_id=users[5] #Based on the popularity 
pm.recommend(user_id) #Here recommender accepts a user id and gives recommended songs for that user.
```

```python
user_id=users[7] #Using popularity model for making predictions on a particular user. 
pm.recommend(user_id)
```

```python
#SONG RECOEMMENDER WITH PERSONALIZATION(Item similarity based collaborative filtering model)

#Build a song recommedation with personalization
#item_similarity_recommender_py-Class of item similarity based personalization is used as a black box
is_model=Recommenders.item_similarity_recommender_py() #creating an instance of item_similarity class
is_model.create(train_data,'user_id','song')
```

```python
#Using personalized model for song recommendation.

#Print the songs for the user in training data
user_id = users[5]
user_items = is_model.get_user_items(user_id)#get_user_items-gives unique items for a given user.
#
print("------------------------------------------------------------------------------------")
print("Training data songs for the user userid: %s:" % user_id)
print("------------------------------------------------------------------------------------")

for user_item in user_items:
    print(user_item)

print("----------------------------------------------------------------------")
print("Recommendation process going on:")
print("----------------------------------------------------------------------")

#Recommend songs for the user using personalized model
is_model.recommend(user_id)
```

```python
#For another user
user_id = users[7]
user_items = is_model.get_user_items(user_id)

print("Training data songs for the user userid: %s:" % user_id)
print("------------------------------------------------------------------------------------")

for user_item in user_items:
    print(user_item)

print("----------------------------------------------------------------------")
print("Recommendation process going on:")
print("----------------------------------------------------------------------")

#Recommend songs for the user using personalized model
is_model.recommend(user_id)
```

```python
#We can apply this model to find similar songs to any song in the dataset.
is_model.get_similar_items(['U Smile - Justin Bieber'])
```

```python
#Use of recommender to get similar songs for the following song.
song = 'Yellow - Coldplay'
is_model.get_similar_items([song])
```

```python
#COMPAIRING TWO MODELS.
#i.e Popularity based and collaborative based using Percision-Recall curves.

#Evaluation.precision_recall_calculator-class for calculating Precision and Recall. 

#Define what percentage of users to use for precision recall calculation
user_sample = 0.05

#Instantiate the precision_recall_calculator class
pr = Evaluation.precision_recall_calculator(test_data, train_data, pm, is_model)

#Call method to calculate precision and recall values
(pm_avg_precision_list, pm_avg_recall_list, ism_avg_precision_list, ism_avg_recall_list) = pr.calculate_measures(user_sample)
```

```python
#Code to plot precision recall curve.
import pylab as pl #Funvctionality of Python + numpy

#Method to generate precision and recall curve
def plot_precision_recall(m1_precision_list, m1_recall_list, m1_label, m2_precision_list, m2_recall_list, m2_label):
    pl.clf()    #Cleraing previous figure.
    pl.plot(m1_recall_list, m1_precision_list, label=m1_label)
    pl.plot(m2_recall_list, m2_precision_list, label=m2_label)
    pl.xlabel('Recall')
    pl.ylabel('Precision')
    pl.ylim([0.0, 0.20])
    pl.xlim([0.0, 0.20])
    pl.title('Precision-Recall curve')
    #pl.legend(loc="upper right")
    pl.legend(loc=9, bbox_to_anchor=(0.5, -0.2))
    pl.show()
```

```python
#Generate precision recall curve using pickled results on a larger datset(subset)(#Python 3).
print("Plotting precision recall curves for a larger subset of data (100,000 rows) (user sample = 0.005).")

#Read the generated files. 
pm_avg_precision_list = joblib.load('pm_avg_precision_list_3.pkl')
pm_avg_recall_list = joblib.load('pm_avg_recall_list_3.pkl')
ism_avg_precision_list = joblib.load('ism_avg_precision_list_3.pkl')
ism_avg_recall_list = joblib.load('ism_avg_recall_list_3.pkl')

print("Plotting precision recall curves.")
plot_precision_recall(pm_avg_precision_list, pm_avg_recall_list, "popularity_model",
                      ism_avg_precision_list, ism_avg_recall_list, "item_similarity_model")
```
